#### My original Tableau Dashboard contained citi bike data for NYC and Jersey City. However, it was over 17 million rows meaning it exceded the Tableau Public limit of 15 million rows. This notebook shows how I whittled the data down so that I could publish the dashboard on Tableau Public

```python
import pandas as pd
import glob
```

```python
#make a list of all the Jersey City csvs
jc_filenames_list = glob.glob('Resources/JC*.csv')
jc_filenames_list
```

```python
#read in each csv into a df
jc_dfs_list = [pd.read_csv(filename) for filename in jc_filenames_list]
```

```python
#make a list of all the NYC csvs
ny_filenames_list = glob.glob('Resources/20*.csv')
ny_filenames_list
```

```python
#read in each csv into a df
ny_dfs_list = [pd.read_csv(filename) for filename in ny_filenames_list]
```

```python
#add a column for the areas to the dfs
for dataframe in jc_dfs_list:
    dataframe['Area'] = 'Jersey City'
```

```python
#add a column for the areas to the dfs
for dataframe in ny_dfs_list:
    dataframe['Area'] = 'NYC'
```

```python
#concactenate all the jc dfs
all_jc_df = pd.concat(jc_dfs_list, ignore_index=True)
all_jc_df
```

```python
#concactenate all the nyc dfs
all_ny_df = pd.concat(ny_dfs_list, ignore_index=True)
all_ny_df
```

```python
#concactenate all the dfs
nyjc_df = pd.concat([all_ny_df, all_jc_df], ignore_index=True)
nyjc_df
```

```python
#drop the name_localizedValue0 column
nyjc_df.drop(['name_localizedValue0'], axis=1, inplace=True)
```

```python
nyjc_df.shape
```

```python
#drop any rows that contain null values
nyjc_df.dropna(inplace=True)
nyjc_df.shape
```

#### Down to 16 mil rows...just need to drop 'a few' more

```python
#drop all longitudes with 0 as the value
nyjc_df = nyjc_df[nyjc_df['end station longitude'] != 0]
nyjc_df = nyjc_df[nyjc_df['start station longitude'] != 0]
nyjc_df.shape
```

```python
#something's off about the tripduration so sort by that
nyjc_df.sort_values(by='tripduration', ascending=False)
```

#### For the sake of size, I'm going to remove those outliers (all rows where the trip duration is longer than 1 day). Clearly, nobody is riding the bike for that many hours.

```python
#delete rows where trip duration is greater than 1 day (1440 mins)
nyjc_df = nyjc_df[nyjc_df['tripduration'] < 1441]
nyjc_df.shape
```
