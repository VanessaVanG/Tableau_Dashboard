```python
import pandas as pd
import glob
```

```python
#make a list of all the Jersey City csvs
jc_filenames_list = glob.glob('Resources/JC*.csv')
jc_filenames_list
```

```python
#read in each csv into a df
jc_dfs_list = [pd.read_csv(filename) for filename in jc_filenames_list]
```

```python
#make a list of all the NYC csvs
ny_filenames_list = glob.glob('Resources/20*.csv')
ny_filenames_list
```

```python
#read in each csv into a df
ny_dfs_list = [pd.read_csv(filename) for filename in ny_filenames_list]
```

```python
#add a column for the areas to the dfs
for dataframe in jc_dfs_list:
    dataframe['Area'] = 'Jersey City'
```

```python
jc_dfs_list[0]
```

```python
#add a column for the areas to the dfs
for dataframe in ny_dfs_list:
    dataframe['Area'] = 'NYC'
```

```python
ny_dfs_list[0]
```

```python
#concactenate all the jc dfs
all_jc_df = pd.concat(jc_dfs_list, ignore_index=True)
all_jc_df
```

```python
#concactenate all the nyc dfs
all_ny_df = pd.concat(ny_dfs_list, ignore_index=True)
all_ny_df
```

```python
#concactenate all the dfs
nyjc_df = pd.concat([all_ny_df, all_jc_df], ignore_index=True)
nyjc_df
```

```python
#export as csv
nyjc_df.to_csv('nyjc.csv')
```
